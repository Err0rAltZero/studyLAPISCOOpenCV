# Filtros Passa-Baixa

Pode-se pensar na frequência como a taxa de mudança em uma imagem. Bordas e outras transições de forte intensidade (como ruído) em uma imagem contribuem significantemente para o conteúdo de alta frequência de sua transformada de Fourier. Consequentemente, a suavização (borramento, *blur*) é obtida no domínio da frequência pela atenuação de alta frequência. Ou seja, por filtragem de passa-baixa (*lowpass*) (Gonzalez, 2008).

Filtro passa-faixa é um dispositivo que permite a passagem de frequências de certa faixa e rejeita, ou atenua, outras frequências fora dessa faixa. Assim, um filtro passa-baixa é um dispositivo que permite a passagem de baixas frequências e atenua a amplitude de frequência superiores à frequência de corte estabelecida.

```python
#Inicialização
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
import math as m
from IPython.display import Image
```

```python
#Imagens básicas como entrada
img00 = cv.imread("../gcPDI/imgSimple/00sDot.png",0)
img01 = cv.imread("../gcPDI/imgSimple/01sHorizontal.png",0)
img02 = cv.imread("../gcPDI/imgSimple/02sVertical.png",0)
img03 = cv.imread("../gcPDI/imgSimple/03sX.png",0)
img04 = cv.imread("../gcPDI/imgSimple/04sSquare.png",0)
img05 = cv.imread("../gcPDI/imgSimple/05sCircle.png",0)
img10 = cv.imread("../gcPDI/imgSimple/10mDots.png",0)
img11 = cv.imread("../gcPDI/imgSimple/11mHorizontals.png",0)
img12 = cv.imread("../gcPDI/imgSimple/12mHorizontals2.png",0)
img13 = cv.imread("../gcPDI/imgSimple/13mVerticals.png",0)
img14 = cv.imread("../gcPDI/imgSimple/14mVerticals2.png",0)
img15 = cv.imread("../gcPDI/imgSimple/15mSquares.png",0)
img16 = cv.imread("../gcPDI/imgSimple/16mCircles.png",0)
img20 = cv.imread("../gcPDI/imgSimple/20random.png",0)
img21 = cv.imread("../gcPDI/imgSimple/21colorRandom.png",0)
simpleImg = [img00, img01, img02, img03, img04, img05, img10, img11, img12, img13, img14, img15, img16, img20, img21]              
for i in range(15):
    plt.subplot(3,5,i+1)
    plt.imshow(simpleImg[i],cmap='gray')
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#Transformada de Fourier das entradas
for i in range(15):
    f = np.fft.fft2(simpleImg[i])
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20*np.log(np.abs(fshift))
    plt.subplot(3,5,i+1)
    plt.imshow(magnitude_spectrum,cmap='gray')
    plt.xticks([]),plt.yticks([])
plt.show()
```
