### Teoria

Equalização de Histograma é um método que melhora o contraste de uma imagem, a fim de esticar a faixa de intensidade. Tal técnicas baseia-se na redistribuição dos valores de intensidade da imagem na escala de cinza.

A aplicação de consiste em uma transformação($T$) na imagem ($g(x,y)$) da seguinte forma:

$$ h(x,y) = T[g(x,y)] $$

### Manipulações

### Desenvolvimento do Cálculo do Histograma Equalizado

($h(x,y)$) é a imagem após o processo de normalização. Tal transformação($T$) pode ser expressa da seguinte forma:

$$ h(i,j) = \frac{g(i,j) - min(g(x,y)}{max(g(x,y) - min(g(x,y))} \times Imax $$

$$ i = 0,1,2...M $$

$$ j = 0,1,2...N $$

Sendo, $h(i,j)$ o novo pixel da imagem, $g(i,j)$ o pixel da imagem original, $min(g(x,y))$ e $max(g(x,y))$ o mínimo e máximo valor de intensidade na imagem respectivamente, com $i$ e $j$ variando até o tamanho da imagem e $Imax$ a intesidade máxima na nova escala adotada.

```python
def histEqual(img):
    I,J = img.shape
    imgH = np.zeros(img.shape)
    nofpix=I*J
    freq = np.zeros((256,1))
    probf= np.zeros((256,1))
    probc = np.zeros((256,1))
    csum = np.zeros((256,1))
    output = np.zeros((256,1))
    print("#",end=" ")
    for i in range(I):
        for j in range(J):
            value=img[i][j]
            freq[value]=freq[value]+1
            probf[value]=freq[value]/nofpix
    nsum=0
    no_bins=255
    #Probabilidade de distribuição acumulativa calculada
    for i in range(256):
        nsum=nsum+freq[i]
        csum[i]=nsum
        probc[i]=csum[i]/nofpix
        temp = np.uint8(probc[i]*no_bins)
        output[i] = temp
    for i in range(I):
        for j in range(J):
            imgH[i][j]=output[img[i][j]]
    return np.uint8(imgH)
```

```python
def calcHist (Image):
    ### Imports
    import numpy as np
    import matplotlib.pyplot as plt
    import math as m
    import navFunc as nf
    # Load image into numpy matrix
    A = Image
    size = nf.structtype()
    size.A = nf.structtype()
    size.A.lin, size.A.col = A.shape
    #################### Calculate Histogram
    ## Pre-setes:
    if A.dtype == 'uint8':
        #buffer = np.linspace(0, 255, num = 2**8)
        #buffer = np.zeros((A.max()+1) - (A.min()))
        buffer = np.zeros((256))
    ## Read the intire matrix element-by-element:
    for j in range((0), size.A.lin):
        for k in range((0), size.A.col):
            buffer[(A[j, k])] += 1
    print('C',end="_")
    return 100*buffer/(size.A.lin*size.A.col)
```

```python
#OpenCV
#Início: Script de plotagem
f1, axarr1 = plt.subplots(3, 3, constrained_layout=True)
f2, axarr2 = plt.subplots(3, 3, constrained_layout=True)
f1.set_figheight(10)
f1.set_figwidth(10)
f2.set_figheight(10)
f2.set_figwidth(10) 
#Matrix[i][j]
for i in range(3):
    for j in range(3):
        img = histEqual(imgMatrix[i][j])
        axarr1[i,j].imshow(img, vmin=0, vmax=255, cmap='gray')
        axarr1[i,j].set_title(titleMatrix[i][j])
        hist = calcHist(img)
        axarr2[i,j].stem(hist,linefmt='C0--',markerfmt='C3*')
        axarr2[i,j].set_title(titleMatrix[i][j])      
f1.suptitle("Imagens equalizadas pelo histEqual()",fontsize=18)
f2.suptitle("Histogramas pelo histEqual() + matplotlib",fontsize=18)
plt.show()
#Final
```
