### Teoria

Multilimiarização é um processo que segmenta uma imagem em nível de cinza em várias regiões distintas. Este processo determina mais de um limiar para a imagem dada e segmenta a imagem em certas regiões de intensidade de cinza.

Como OpenCV não possui implementações que favorecam a utilização de vários níveis, é mais prático fazer uma implementação própria.

### Desenvolvimento da Limiarização

A transformação pode ser exemplificada pelo seguinte conjunto de equações: 

$g(x,y) = 0 $, se $f(x,y) < Linf$, 

ou $g(x,y) = h$, se $Linf<g(x,y)<Lsup$, 

ou $g(x,y) = max$, se $f(x,y) \geq Lsup$

O resultado é uma imagem com valores discretos, distintos entre o máximo e o mínimo da intensidade da imagem.

```python
def multiLimiar (img, valL, valT):
    #valL: valores Limites 
    #valT: valores Troca (pixel recebe esse valor)
    #Por conta do while not, a função aceita vários números de regiões.
    if len(valL)!=len(valT)-1:
        print("Erro! Vetores Incompatíveis")
        return 0
    lmax = len(valL)-1
    tmax = len(valT)-1
    valL.sort() #Organiza o vetor
    #valT.sort() #Organiza o vetor
    I,J = img.shape #Recebe o número de linhas e colunas
    D = np.zeros(img.shape) #Cria uma matriz com as dimensões da img recebida
    for i in range(I):
        for j in range(J):
            #if Filter.multiLimiar.size == 3:
            if img[i, j] > valL[lmax]:
                D[i, j] = valT[tmax]
            else:
                aux=0
                while not img[i,j] <= valL[aux]: #Enquanto o pixel não for <= do que o valor da posição aux do vetor
                    aux+=1 #Incrementa aux, para testar o próximo valor
                D[i,j] = valT[aux]
    D = np.uint8(D)
    print('ML',end="#")
    return D
```

```python
vL = [77,164] #Região 1: de 0 até 77, Região 2: de 78 até 164, Região 3: de 164 até 255
vT = [0,127,255] #Preto, cinza, branco
dst = multiLimiar(imgB,vL,vT)
#Início: Script de plotagem
f, axarr = plt.subplots(1, 2, constrained_layout=True)
#Matrix[i][j]
axarr[0].imshow(imgB,"gray")
hist = cv.calcHist([imgB],[0],None,[256],[0,256])
axarr[1].stem(hist,linefmt='C0--',markerfmt='C3*')
plt.show()
#Final
#Início: Script de plotagem
f, axarr = plt.subplots(1, 2, constrained_layout=True)
#Matrix[i][j]
axarr[0].imshow(dst,"gray")
hist = cv.calcHist([dst],[0],None,[256],[0,256])
axarr[1].stem(hist,linefmt='C0--',markerfmt='C3*')
plt.show()
#Final
```
