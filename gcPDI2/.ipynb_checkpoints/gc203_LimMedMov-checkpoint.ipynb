# Morfologia Matemática (Erosão e Dilatação)

## Introdução

As **Transformações Morfológicas** são algumas operações simples baseadas na forma da imagem. Ele precisa de duas entradas, a imagem original e o elemento estruturante, ou *kernel*, que decide a natureza da operação.

Baseado em Gonzalez e Woods (2008), e Sonka, Hlavac e Boyle (2015); a linguagem da **Morfologia Matemática** é a teoria dos conjuntos. Conjuntos (*sets*) em **Morfologia Matemática** representam objetos em uma imagem. Assim, a **morfologia matemática** usa ferramentas de álgebra não linear e opera com conjuntos de pontos, sua conectividade e forma. As operações morfológicas simplificam as imagens, quantificam, e preservam as principais características da(s) forma(s) dos objetos.

As operações morfológicas são usadas predominantemente para:
+ Pré-processamento de imagem (filtragem de ruído, simplificação de forma).
+ Aprimoramento da estrutura do objeto (esqueletização, afinamento, espessamento, casco convexo).
+ Segmentar objetos do fundo.
+ Descrição quantitativa de objetos (área, perímetro, projeções,).

**Erosão** e **Dilatação** são operadores considerados elementares, pois a maior parte dos operadores morfológicos utilizados podem ser implementados a partir dos mesmos.

## Implementação

```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
import math
from IPython.display import Image

kernel = np.ones((3,3))

imgA = cv.imread("../gcPDI2/imgPDI2/gcmyshapes.tif",0)
plt.imshow(imgA,"gray")
plt.show()

f, axarr = plt.subplots(1, 3, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)
axarr[0].imshow(cv.erode(imgA,kernel,iterations = 1),'gray')
axarr[1].imshow(imgA,'gray')
axarr[2].imshow(cv.dilate(imgA,kernel,iterations = 1), 'gray')
plt.show()
```

Ideia: Usar uma matriz binária, onde o conjunto de valores unitários significasse a máscara em si, isso é, uma forma (1) contida em uma matriz (0).
Para Erosão, caso a máscara (1) toque todo os pontos brancos (255), a soma dos valores contidos nessa matriz deve ser igual ao valor da soma desses valores unitários (*kernel.sum()*) multiplicado por 255 (\**255*), para conservar o ponto como branco:

+ ***if tempsum == kernel.sum()\*255:***

Para Dilatação, caso a máscara (1) toque algum ponto branco (255), basta que a soma dos valores contidos nessa matriz seja diferente de zero.

+ ***if tempsum != 0:***

### Erosão

```python
#Erosion
def mmErosion(img, kernel, mode=''):
    Arow, Acol = img.shape
    Brow, Bcol = kernel.shape
    
    ROWpad = math.floor((Brow/2))
    COLpad = math.floor((Bcol/2))

    Crow = Arow + ROWpad*2 
    Ccol = Acol + COLpad*2
    C = np.zeros((Crow,Ccol))    
    C[(ROWpad):(Crow-ROWpad), (COLpad):(Ccol-COLpad)] = img.copy()
    
    if mode == 'FULL':
        plt.imshow(C,'gray')
        plt.title("Imagem com Bordas: {}x{}".format(Crow,Ccol))
        plt.show()

    D = np.zeros(img.shape)
    for Ci in range(0, Arow): 
        for Cj in range(0, Acol):
            tempsum = 0
            for Ki in range(0, Brow):
                for Kj in range(0, Bcol):
                    tempsum = kernel[Ki,Kj]*C[Ci+Ki,Cj+Kj]+tempsum
            if tempsum == kernel.sum()*255:
                D[Ci,Cj]=255            
    return D
```

```python
plt.imshow(imgA,'gray')
plt.title("Original")
plt.show()

kernel = np.ones((3,3))
plt.imshow(kernel,'gray')
plt.title("Kernel")
plt.show()

result = mmErosion(imgA, kernel, mode='FULL')
plt.imshow(result,'gray')
plt.title("Imagem após Erosão")
plt.show()
```

### Dilatação

```python
#Dilation
def mmDilation(img, kernel, mode=''):
    Arow, Acol = img.shape
    Brow, Bcol = kernel.shape
    
    ROWpad = math.floor((Brow/2))
    COLpad = math.floor((Bcol/2))

    Crow = Arow + ROWpad*2 
    Ccol = Acol + COLpad*2
    C = np.zeros((Crow,Ccol))    
    C[(ROWpad):(Crow-ROWpad), (COLpad):(Ccol-COLpad)] = img.copy()

    if mode == 'FULL':
        plt.imshow(C,'gray')
        plt.title("Imagem com Bordas: {}x{}".format(Crow,Ccol))
        plt.show()
        
    D = np.zeros(img.shape)
    for Ci in range(0, Arow): 
        for Cj in range(0, Acol):
            tempsum = 0
            for Ki in range(0, Brow):
                for Kj in range(0, Bcol):
                    tempsum = kernel[Ki,Kj]*C[Ci+Ki,Cj+Kj]+tempsum
            if tempsum != 0:
                D[Ci,Cj]=255            
    return D                    
```

```python
plt.imshow(imgA,'gray')
plt.title("Original")
plt.show()

kernel = np.ones((3,3))
plt.imshow(kernel,'gray')
plt.title("Kernel")
plt.show()

result = mmDilation(imgA, kernel, mode='FULL')
plt.imshow(result,'gray')
plt.title("Imagem após Dilatação")
plt.show()
```

## Conclusão

```python
imgA = cv.imread("../gcPDI2/imgPDI2/gcmyshapes.tif",0)
plt.imshow(imgA,"gray")
plt.show()
```

```python
f, axarr = plt.subplots(1, 3, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)

kernel = np.ones((3,3))
result = mmErosion(imgA, kernel)
axarr[0].imshow(result,'gray')
axarr[0].set_title("3x3")

kernel = np.ones((5,5))
result = mmErosion(imgA, kernel)
axarr[1].imshow(result,'gray')
axarr[1].set_title("5x5")

kernel = np.ones((7,7))
result = mmErosion(imgA, kernel)
axarr[2].imshow(result,'gray')
axarr[2].set_title("7x7")
plt.show()
```

```python
f, axarr = plt.subplots(1, 3, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)
kernel = np.ones((3,3))
result = mmErosion(imgA, kernel)
for i in range(3):
    axarr[i].imshow(result,'gray')
    axarr[i].set_title("Kernel 3x3\n{} Iterações".format(i+1))
    result = mmErosion(result, kernel)
plt.show()
```

```python
f, axarr = plt.subplots(2, 2, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)

kernel = np.ones((3,9))
result = mmErosion(imgA, kernel)
axarr[0,0].imshow(result,'gray')
axarr[0,0].set_title("3x9")

kernel = np.ones((9,3))
result = mmErosion(imgA, kernel)
axarr[0,1].imshow(result,'gray')
axarr[0,1].set_title("9x3")

kernel = np.zeros((3,3))
kernel[0,0] = 1
kernel[1,1] = 1
kernel[2,2] = 1
result = mmErosion(imgA, kernel)
axarr[1,0].imshow(result,'gray')
axarr[1,0].set_title("\"\\\" como 3x3")

kernel = np.zeros((3,3))
kernel[0,2] = 1
kernel[1,1] = 1
kernel[2,0] = 1
result = mmErosion(imgA, kernel)
axarr[1,1].imshow(result,'gray')
axarr[1,1].set_title("\"/\" como 3x3")

plt.show()
```

```python
f, axarr = plt.subplots(1, 3, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)

kernel = np.ones((3,3))
result = mmDilation(imgA, kernel)
axarr[0].imshow(result,'gray')
axarr[0].set_title("3x3")

kernel = np.ones((5,5))
result = mmDilation(imgA, kernel)
axarr[1].imshow(result,'gray')
axarr[1].set_title("5x5")

kernel = np.ones((7,7))
result = mmDilation(imgA, kernel)
axarr[2].imshow(result,'gray')
axarr[2].set_title("7x7")
plt.show()
```

```python
f, axarr = plt.subplots(1, 3, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)
kernel = np.ones((3,3))
result = mmDilation(imgA, kernel)
for i in range(3):
    axarr[i].imshow(result,'gray')
    axarr[i].set_title("Kernel 3x3\n{} Iterações".format(i+1))
    result = mmDilation(result, kernel)
plt.show()
```

```python
f, axarr = plt.subplots(2, 2, constrained_layout=True)
f.set_figheight(10)
f.set_figwidth(10)

kernel = np.ones((3,9))
result = mmDilation(imgA, kernel)
axarr[0,0].imshow(result,'gray')
axarr[0,0].set_title("3x9")

kernel = np.ones((9,3))
result = mmDilation(imgA, kernel)
axarr[0,1].imshow(result,'gray')
axarr[0,1].set_title("9x3")

kernel = np.zeros((3,3))
kernel[0,0] = 1
kernel[1,1] = 1
kernel[2,2] = 1
result = mmDilation(imgA, kernel)
axarr[1,0].imshow(result,'gray')
axarr[1,0].set_title("\"\\\" como 3x3")

kernel = np.zeros((3,3))
kernel[0,2] = 1
kernel[1,1] = 1
kernel[2,0] = 1
result = mmDilation(imgA, kernel)
axarr[1,1].imshow(result,'gray')
axarr[1,1].set_title("\"/\" como 3x3")

plt.show()
```

Os resultados mostram que iterações sucessivas do filtro unitário quadrado são equivalentes à aplicações de filtros de maior dimensão.

A modificação da máscara en proporção, ou mesmo forma, serve para controlar o sentido, a forma e/ou quais detalhes a saída da operação irá priorizar.
